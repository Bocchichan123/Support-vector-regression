```python
import pandas
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_predict
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import make_scorer
```

```python
dataset = pandas.read_csv('dataset.csv', index_col= 'Year', parse_dates = True)
dataset = dataset.dropna()
print('Shape of Data', dataset.shape)
dataset.head()
```

```python
dataset['Yield'].plot(figsize = (12, 5))
```

```python
from statsmodels.tsa.stattools import adfuller

def ad_test(dataset):
    dftest = adfuller(dataset, autolag = 'AIC')
    print("1. ADF :",dftest[0])
    print("2. P-Value :", dftest[1])
    print("3. Num of Lags :", dftest[2])
    print("4. Num of Observations Used :", dftest[3])
    print("5. Critical Values :")
    for key, val in dftest[4].items():
        print("\t",key, ":", val)
```

```python
ad_test(dataset['Yield'])
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
print(dataset.shape)
train = dataset.iloc[:-11]
test = dataset.iloc[-11:]
print(train.shape, test.shape)
```

```python
train.plot()
test.plot()
```

```python
model = ARIMA(train['Yield'], order = (1,1,0))
model = model.fit()
model.summary()
```

```python
start = len(train)
end = len(train)+len(test)-1
pred = model.predict(start = start, end = end, typ = 'levels')
print(pred)
```

```python
test
```

```python
pred.plot(legend = True)
test['Yield'].plot(legend = True)
```

```python
import numpy as np
print('The RMSE is: %.2f'% np.sqrt(mean_squared_error(pred, test['Yield'])))
```

```python
print('The MAE is: %.2f'% mean_absolute_error(pred, test['Yield']))
```
