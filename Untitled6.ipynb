```python
import pandas as pandas
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pandas.read_csv('dataset.csv', index_col= 'Year', parse_dates = True)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.index.freq = 'YS'
```

```python
dataset.info()
```

```python
dataset.plot(figsize = (12, 5))
```

```python
train = dataset.iloc[:25]
```

```python
test = dataset.iloc[25:]
```

```python
train.plot()
test.plot()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
model = ExponentialSmoothing(train['Yield'], trend = 'mul', seasonal = 'mul', seasonal_periods = 12).fit()
```

```python
test.plot()
model.forecast(11).plot()
```

```python
dataset.plot(figsize = (12,8))
model.forecast(11).plot()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
print('The RMSE is: %.2f'% np.sqrt(mean_squared_error(model.forecast(11), test)))
```

```python
from sklearn.metrics import mean_absolute_error
print('The MAE is: %.2f'% mean_absolute_error(model.forecast(11), test))
```
