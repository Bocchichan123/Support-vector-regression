```python
import pandas as pd
from matplotlib import pyplot
from matplotlib import pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
from pylab import rcParams
rcParams['figure.figsize'] = 20,5
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('dataset.csv', header = 0, index_col = 0)
```

```python
dataset.head(36)
```

```python
plt.plot(dataset[0:36]['Yield'])
plt.show()
```

```python
rollingseries = dataset[0:36].rolling(window = 26)
rollingmean = rollingseries.mean()
print(rollingmean.head(10))
```

```python
rollingmean.plot(color = 'red')
pyplot.show()
```

```python
plt.plot(dataset[0:36]['Yield'])
rollingmean.plot(color = 'red')
plt.show()
```

```python
data = dataset[0:36]
fit1 = SimpleExpSmoothing(data).fit(smoothing_level = 0.2, optimized = False)
fit2 = SimpleExpSmoothing(data).fit(smoothing_level = 0.6, optimized = False)
```

```python
plt.figure(figsize = (18,8))
plt.plot(data, marker = 'o', color = 'black', label = 'Data')
plt.plot(fit2.fittedvalues, marker = 'o', color = 'red', label = '60%')
plt.legend()
plt.title('Exponential Smoothing')
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
print('The RMSE for 0.2 is: %.2f'% np.sqrt(mean_squared_error(fit1.fittedvalues, data['Yield'])))
print('The RMSE for 0.6 is: %.2f'% np.sqrt(mean_squared_error(fit2.fittedvalues, data['Yield'])))
```

```python
from sklearn.metrics import mean_absolute_error
print('The MAE for 0.8 is: %.2f'% mean_absolute_error(fit2.fittedvalues, data['Yield']))
```
